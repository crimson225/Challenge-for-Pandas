```python
#pandas import
import pandas as pd
import numpy as np
#file import and read as csv
startf ="C:/users/steel/GT-ATL-DATA-PT-12-2019-U-C/Homework/04-Pandas/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"
#made into DataFrame
readf = pd.read_csv(startf)
readf.head()    
```

```python
# Total Number of Players
unique_name = len(readf["SN"].unique())
Playerdf = pd.DataFrame({"Total Players": [unique_name]})
Playerdf
```

```python
#Unique Items
unique_item= len(readf["Item ID"].unique())
#Total Revenue
ttl_price = readf["Price"].sum()
# Number of Purchases
ttl_purchase=readf['Purchase ID'].count()
#Average Purchase
avg_purchase = ttl_price/ttl_purchase

totanalysis = pd.DataFrame ({"Number of Unique Items": [unique_item], "Average Price":avg_purchase,
                            "Number of Purchases":ttl_purchase,"Total Revenue":ttl_price})
totanalysis["Average Price"] = totanalysis["Average Price"].astype(float).map("${:.2f}".format)
totanalysis["Total Revenue"] = totanalysis["Total Revenue"].astype(float).map("${:,.2f}".format)

totanalysis
```

```python
genderdf = readf.groupby(['Gender', 'SN'])
playdf = genderdf.size()
gendercount = playdf.groupby(['Gender']).count()
#print(gendercount)
total = gendercount.sum()
#print(total)
pct = (100*gendercount)/total
#print(pct)
gdf= pd.DataFrame ({"Total": gendercount, "Percentage": pct})
gdf["Percentage"] = gdf["Percentage"].map("{:.2f}%".format)
gdf
```

```python
genderp = readf.groupby(['Gender', 'Purchase ID', 'Price'])
pricedf = genderp.size()
pcount = pricedf.groupby(['Gender']).count()
#pcount (counts are correct)
Total = readf.groupby(by='Gender')['Price'].sum()
#Total (totals correct)
avg = Total/pcount
#avg( avg purchase correct)

avgpp = Total/gendercount
#avgpp (avg per person correct)
Total = Total.map("${:,.2f}".format)
avg = avg.map("${:.2f}".format)
avgpp = avgpp.map("${:.2f}".format)
padf = pd.DataFrame({"Purchase Count": pcount, "Average Purchase Price": avg, "Total Purchase Value": Total,
                    "Avg Total Purchase per Person": avgpp})

padf
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]
labels = ["Less than 10", "10 to 14","15 to 19","20 to 24","25 to 29","30 to 34",
          "35 to 39", "Over 40"]
pd.cut(readf["Age"], bins, labels=labels)
readf["Age Groups"] = pd.cut(readf["Age"], bins, labels=labels)
agegrp = readf.groupby(["Age Groups"])
ageitm = agegrp["Item ID"].count()
ageavg = agegrp["Price"].mean() #avg purchase
agetot = agegrp["Price"].sum() #total purchase
ageavg = ageavg.map("${:.2f}".format)
agetot = agetot.map("${:.2f}".format)
agedf = pd.DataFrame({"Purchase Count": ageitm,"Purchase Price": ageavg,"Total Price":agetot})
agedf
```

```python
spendf = readf.groupby(['SN', 'Purchase ID', 'Price'])
spenf = spendf.size()

spdf = spenf.groupby(['SN']).count() #counting purchases (Correct)
#spdcnt = spdf["SN"].count()
spentot = readf.groupby(by='SN').sum()['Price'] #total purchase per SN
spenavg = readf.groupby(by='SN').mean()['Price']#avg per person

spenddf = pd.DataFrame ({"Purchase Counts": spdf, "Average Purchase Price":spenavg, "Total Purchase Value": spentot})

spensort = spenddf.sort_values("Total Purchase Value", ascending=False)
spensort["Average Purchase Price"] = spensort["Average Purchase Price"].map("${:.2f}".format)
spensort["Total Purchase Value"] = spensort["Total Purchase Value"].map("${:.2f}".format)
spensort.head()
#spenddf.head()
```

```python
itmgrp = readf.groupby(["Purchase ID", "Item Name", "Price"])
itmgr = itmgrp.size()
itm = itmgr.groupby(["Item Name"]).count() #items counted (correct)

itmttl = readf.groupby(by='Item Name')['Price'].sum() #total purchse value (correct)
itmavg = readf.groupby(by='Item Name')['Price'].mean()
itmavg = itmavg.map("${:.2f}".format)

itmdf = pd.DataFrame({"Purchase Count": itm,"Item Price": itmavg, "Total Purchase Value": itmttl})
itmdfsort = itmdf.sort_values("Purchase Count", ascending = False)
itmdfsort["Total Purchase Value"] = itmdfsort["Total Purchase Value"].map("${:.2f}".format)

itmdfsort.head(5)
```
